```python
from sklearn import tree
from sklearn import datasets
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = datasets.load_digits() 
```

```python
print(data.keys())
for f in data.keys():
    t = type(data[f])
    print('key: {}, type: {}'.format(f, t))
    if t == np.ndarray:
        print('shape: {}'.format(data[f].shape))
```

```python
df = pd.DataFrame(data=data.data)
df['target'] = data.target
print(data.target_names)
```

```python
plt.imshow(data.data[100].reshape(8, 8).astype(np.int32), cmap='gray') 
```

```python
X_train = data.data
y_train = data.target
```

```python
model = tree.DecisionTreeClassifier(max_depth=5) 
```

```python
model.fit(X_train, y_train) 
```

```python
y = model.predict(X_train) 
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y_train)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_train[:,0], X_train[:,1], X_train[:,2], c = y)
plt.show()
```

```python
print(model.score(X_train, y_train)) 
```
